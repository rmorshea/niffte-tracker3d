```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def getdat(textfile):
    f=open(textfile,'r')
    flist=[]
    f.readline()
    line=f.readline()
    while len(line)!=0:
        line=[int(i) for i in line.split(' ')]
        flist.append(line)
        line=f.readline()
    return flist
```

```python
flist=getdat('event31.txt')
padDict={}
xs=[]
ys=[]
zs=[]
for v in flist:
    xs.append(v[1]);ys.append(v[2]);zs.append(v[3])
    padId=str(v[1])+str(v[2])
    if padId not in padDict:
        padDict[padId]=[(v[-2],v[-1])]
    else:
        padDict[padId].append((v[-2],v[-1]))
print padDict

```

```python
fig = plt.figure(1, figsize=(10,8))
ax = fig.add_subplot(1,1,1, projection = '3d')
ax.scatter(ys,xs,zs)
```

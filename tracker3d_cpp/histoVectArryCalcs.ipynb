```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def getdat(textfile):
    f=open(textfile,'r')
    flist=[]
    line=f.readline()
    while len(line)!=0:
        line=[int(i) for i in line.split('\t')]
        flist.append(line)
        line=f.readline()
    return flist
```

```python
vect_arry = getdat("sifterTimes.txt")
vect_arry = zip(*vect_arry)
```

```python
plt.figure(figsize=(12,6))
plt.hist(vect_arry[0],bins=50)
plt.title('Vector Sifting')
plt.xlabel('calculation time ($\mu$s)',fontsize=13)
plt.ylabel('count',fontsize=13)
plt.xlim(0,600)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.hist(vect_arry[1],bins=50)
plt.title('Array Sifting')
plt.xlabel('calculation time ($\mu$s)',fontsize=13)
plt.ylabel('count',fontsize=13)
plt.xlim(0,160)
plt.show()
```
